```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

load the image

```python
img = cv2.imread("web-team.jpg")
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.Figure(figsize=(6,4))
plt.imshow(img_rgb)
plt.title("1.Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.Figure(figsize=(6,4))
plt.imshow(gray, cmap ='gray')
plt.title("2. Grayscale Image")
plt.axis("off")
plt.show()
```

```python
resize = cv2.resize(img,(300,200))
cv2.imwrite('resized-clone.jpg',resize)
resized_rgb = cv2.cvtColor(resize,cv2.COLOR_BGR2RGB)
plt.Figure(figsize=(6,4))
plt.imshow(img_rgb)
plt.title("3. Resized Image")
plt.axis("off")
plt.show()
```

```python
edge = cv2.Canny(img,100,200)
plt.Figure(figsize=(6,4))
plt.imshow(edge,cmap='gray')
plt.title("4. Canny Edge detection")
plt.axis("off")
plt.show()
```

```python
img_shape = img.copy()
cv2.rectangle(img_shape,(50,50),(150,150),(0,255,0),3)
cv2.circle(img_shape,(200,200),40,(255,0,0),-1)
cv2.putText(img_shape,"Hello",(10,400),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
shape_rgb = cv2.cvtColor(img_shape,cv2.COLOR_BGR2RGB)
plt.Figure(figsize=(6,4))
plt.imshow(shape_rgb)
plt.title("5. Shape and Text")
plt.axis("off")
plt.show()
```
